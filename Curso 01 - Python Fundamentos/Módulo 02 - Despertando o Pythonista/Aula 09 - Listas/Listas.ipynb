# üß† Fun√ß√µes para listas

### `Append( )`

Usada para adicionar um novo elemento no final de uma lista.

```python
animais = ['cachorro', 'gato', 'passarinho']
```

```python
animais.append('peixe')
animais
```

Podemos adicionar muitos elementos com maior agilidade utilizando uma estrutura de repeti√ß√£o, como o <font color = "orange"><b>`for( )`</font></b>:

```python
numeros = []

for number in range(15):
    numeros.append(number)

print(numeros)
```

E voc√™ tamb√©m pode chamar `number` de uma vari√°vel qualquer, como `i`:

```python
numeros = []

for i in range(15):
  numeros.append(i)

print(numeros)
```

### `Insert( )`

Com ela, pode-se adicionar um novo elemento com uma **posi√ß√£o espec√≠fica**. Neste caso, a fun√ß√£o, al√©m de receber o elemento a ser inserido, recebe o √≠ndice que deve ser dado ao elemento.

```python
numeros.insert(5, 'blabla')  #Aqui o "0" representa o √≠ndice e o "-1" representa o elemento a ser adicionado.
numeros
```

### `Pop( )`

Serve para retirar elementos da lista. Se precisemos retirar um elemento espec√≠fico podemos indicar o √≠ndice do elemento a ser retirado, caso o √≠ndice n√£o seja especificado ser√° retirado o √∫ltimo elemento da lista

```python
animais
```

```python
animais.pop()
animais
```

```python
animais.pop(1)  #Aqui retiramos o elemento com o √≠ndice "1", nesse caso era o "gato".
animais
```

### `Remove( )`

Quando n√£o √© conhecido o √≠ndice do elemento, pode-se retir√°-lo a partir do seu valor. Assim, vai ser retirado o primeiro item que corresponder com o valor recebido.

```python
animais
```

```python
animais.remove('cachorro')
animais
```

### `Reverse( )`

√â usada para deixar a fun√ß√£o "ao contr√°rio". O √∫ltimo item vai virar o primeiro, e o primeiro vai virar o √∫ltimo.


```python
uma_lista
```

```python
uma_lista.reverse()
uma_lista
```

###`Count( )`

Essa fun√ß√£o conta a quantidade de vezes que um elemento aparece na lista em quest√£o.

```python
outra_lista = [1, 1, 2, 2, 2, 3, 3, 3, 3]
```

# üß† Listas e Strings

Apesar de <font color = "orange"><b>`strings`</b></font> serem imut√°veis, podemos criar listas a partir de <font color = "orange"><b>`strings`</b></font> e vice versa.

```python
curso = 'Engenharia de Produ√ß√£o'
```

```python
curso.split()
```

No exemplo anterior a fun√ß√£o <font color = "orange"><b>`split()`</b></font> foi usada. Ela retorna as palavras da <font color = "orange"><b>`string`</b></font> separadas em elementos de uma lista. Como voc√™ pode notar, nada foi especificado entre os par√™nteses, por isso, o Python entende que deve utilizar como m√©todo de separa√ß√£o os espa√ßos em branco entre as palavras. Voc√™ pode escolher o m√©todo de separa√ß√£o, por exemplo v√≠rgulas, pontos, etc.

```python
'mec√¢nica;civil;el√©trica'.split(';') #Aqui especificamos o m√©todo de separa√ß√£o, os ";"
```

Podemos fazer o contr√°rio tamb√©m, ou seja, juntar uma lista em <font color = "orange"><b>`string`</b></font>. Vamos usar a fun√ß√£o<font color = "orange"><b> `join()`</b></font>.

```python
habilitacoes = ['mec√¢nica', 'civil', 'el√©trica']
```

```python
' '.join(habilitacoes)  #Estamos definindo um espa√ßo em branco para conectar os elementos da lista.
```

```python
' e '.join(habilitacoes)
```

```python
print('\n'.join(habilitacoes))
```

#üîÅ Mutabilidade

Diferente da <font color = "orange"><b>`string`</b></font>, a lista √© um tipo de **objeto mut√°vel**.

Lembra do m√≥dulo passado, quando aprendemos que uma <font color = "orange"><b>`string`</b></font> n√£o pode ser mudada? D√° uma olhadinha nesse exemplo:

```python
mensagem = 'BOM DIA!'     #Atribuimos essa string a vari√°vel "mensagem"
print(mensagem)

print(mensagem.lower())  #Lembre-se: quando aplicamos uma fun√ß√£o a uma string, ela retorna uma NOVA string.

print(mensagem)

mensagem = mensagem.lower()   #Portanto, faz sentido reatribuir a vari√°vel "mensagem" uma nova string!
print(mensagem)
```

Agora, vamos ver a diferen√ßa entre <font color = "orange"><b>`listas`</b></font> e <font color = "orange"><b>`strings`</b></font>.

Listas s√£o MUT√ÅVEIS! Como vimos nesse m√≥dulo, **podemos alterar uma lista** usando as fun√ß√µes que aprendemos! Veja o mesmo exemplo usando uma lista:

```python
mensagem = ['Bom']  #Aqui temos nossa lista original, contendo apenas um elemento.
print(mensagem)

mensagem.append('dia') #Usamos a fun√ß√£o "append" para acrescentar um novo elemento a MESMA LISTA, sem criar uma nova lista!
print(mensagem)
```

Agora veja esse outro caso:

```python
mensagem = mensagem.append('brasil')  #Portanto, n√£o faz sentido querer atribuir a vari√°vel a lista que j√° temos.
print(mensagem)
```

Isso modifica a lista, mas depois voc√™ atribui o valor None √† vari√°vel mensagem.

```python
#mensagem.append('brasil')  # modifica a lista E retorna None
#mensagem = None            # voc√™ est√° dizendo: mensagem agora √© igual a None
```

Mesmo que 'brasil' tenha sido adicionado internamente √† lista, voc√™ perde o acesso √† lista original. Agora a vari√°vel mensagem aponta para None.
