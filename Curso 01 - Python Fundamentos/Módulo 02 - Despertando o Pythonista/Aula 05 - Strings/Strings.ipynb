Se quisermos uma parte da <font color = "orange"><b>`string`</font></b>:

**string[in√≠cio:fim:passo]**

* in√≠cio: onde come√ßa (inclusivo)
* fim: onde para (exclusivo)
* passo: de quanto em quanto (pode ser negativo)

```python
#Vai do in√≠cio da string at√© o √≠ndice 5 (exclusivo). Ou seja, pega os caracteres de √≠ndice 0 a 4:

nome[:5]
```

```python
#Esse √© um truque para inverter a string.
#::-1 significa: percorra do fim para o come√ßo, pulando de -1 em -1.

nome[::-1]
```

```python
#Vai do √≠ndice 1 at√© o final da string.
#Come√ßa em a (√≠ndice 1), depois segue normalmente.

nome[1:]
```

```python
#Essa √© uma fatia invertida.
#Come√ßa em √≠ndice 5 (' ' espa√ßo) e vai at√© o √≠ndice 1 (exclusivo), pulando de -1 em -1.
#Ou seja, l√™ de tr√°s pra frente os caracteres de √≠ndice 5 a 2:

nome[5:1:-1]
```

# ‚úÖ Imutabilidade

√â importante lembrar que os objetos do tipo <font color = "orange"><b>`string`</font></b> s√£o imut√°veis, ou seja, n√£o podemos alter√°-los! Talvez voc√™ me conteste dizendo que j√° mudamos uma <font color = "orange"><b>`string`</font></b> antes (quando colocamos elas em letras mai√∫sculas, por exemplo).

Na verdade n√≥s n√£o as alteramos, apenas ordenamos ao programa que as imprimissem de outra maneira. O que acontece √© que elas ter√£o **endere√ßos diferentes**, ou seja, o Python cria um novo objeto (a <font color = "orange"><b>`string`</font></b> impressa do jeito que voc√™ mandou) e os diferencia ocupando dois espa√ßos diferentes na mem√≥ria.

```python
uma_string = 'oi, tudo bem?'
```

```python
id(uma_string)  #Esse √© o n√∫mero de identifica√ß√£o desse objeto para o programa.
```

```python
uma_string.upper()
```

```python
id(uma_string.upper())  #Percebeu que n√£o √© o mesmo n√∫mero? Portanto, n√£o s√£o o mesmo objeto!
```

```python
uma_string == uma_string.upper()
```

Como falamos agora, n√£o podemos mudar uma <font color = "orange"><b>`string`</font></b>, portanto, olha o que acontece se quisermos mudar um car√°cter dela:

```python
uma_string[0]
```

```python
uma_string[0] = 'P'
```
