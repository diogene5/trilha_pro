# üß™ Operadores L√≥gicos

Em Python existem 3 operadores l√≥gicos:  <font color='orange'> **`and`** </font>, <font color='orange'> **`or`** </font> e <font color='orange'> **`not`**</font>.

Exemplo: Considere dois valores booleanos A e B.



*  Operador <font color='orange'> **`and`** </font> ("e"): resultar√° em  <font color='orange'>**`True`**</font>  somente quando **ambos** A e B forem verdadeiros, caso contr√°rio resultar√° em  <font color='orange'>**`False`**</font>.

```python
a = (1 < 2)  # Neste exemplo, ambas express√µes a e b s√£o verdadeiras
b = (2 == 2)

a and b
```

```python
a = (1 < 2)  # Aqui a √© verdadeira, por√©m b n√£o √©, portanto, o resultado de a e b √© falso.
b = (2 == 3)

a and b
```

#üßÆ Estrutura b√°sica: IF

Quando come√ßamos a programar, um recurso b√°sico que precisamos aprender s√£o as **estruturas de condi√ß√£o**. Com elas podemos estipular que um c√≥digo somente seja executado se uma determinada condi√ß√£o for satisfeita.

O comando <font color='orange'> **`if`** </font> executa uma sequ√™ncia de express√µes **apenas se** uma determinada condi√ß√£o for satisfeita.

```python
idade = 20
```

####Exemplo 1

```python
if idade >= 18:
  print("Maior de idade!")
  print("eba!")
```

####Exemplo 2

```python
idade = 10
if idade >= 18:
    print("Maior de idade!")
    print("eba!")
```

Como voc√™ pode ver nesse exemplo, a condi√ß√£o estabelecida para o c√≥digo ser executado √©: a idade ser maior ou igual a 18.

A estrutura <font color='orange'>**`if`**</font> no Python √© seguida pela condi√ß√£o l√≥gica estabelecida e por dois pontos (:). As linhas abaixo dela s√£o as instru√ß√µes que ser√£o executadas caso a condi√ß√£o anterior seja antendida.

***
<font color="grey">

<font color="grey">Aqui vai um conceito muito importante para suas estruturas de condi√ß√£o ficarem certinhas! O Python espera dos seus programadores um pouquinho de organiza√ß√£o em seus c√≥digos, assim, como voc√™ viu no exemplo acima, o comando `if` (e muitos outros) segue um padr√£o de espa√ßamento!

<font color="grey">Os espa√ßamentos diferentes referenciam diferentes blocos. Ou seja, delimitam o come√ßo e o fim de um comando (para aqueles que conhecem o C++, √© como o famoso "abre e fecha chaves").
  
<font color="grey">No exemplo acima, a instru√ß√£o `print` est√° dentro do comando `if` (ela somente ser√° executada se a condi√ß√£o for verdadeira), e por isso recebe o **recuo de quatro espa√ßos**!

J√° quando sua identa√ß√£o √© mais √† frente, ele √© considerado um comando do <font color='orange'>**`if`**</font>, e s√≥ vai acontecer se a condi√ß√£o avaliada for <font color='orange'>**`True`**</font>

#üßÆ Estrutura b√°sica: Else

Nos exemplos acima, t√≠nhamos uma reposta somente quando a condi√ß√£o estipulada era atendida. Quando n√£o era, nada era impresso. Ent√£o, se a idade fosse menor que 18 o programa simplesmente n√£o te respondia. Mas bem que ele podia, n√©?

Para isso, √© preciso que voc√™ estipule o que ele deve responder com o comando <font color='orange'>**`else`**</font>.

```python
idade = 10
if idade >= 18:
  print("Maior de idade!")
else:
  print ("Menor de idade!")
```

O comando <font color='orange'>**`else`**</font> serve para todas as outras possibilidades que n√£o atendem a condi√ß√£o estabelecida no <font color='orange'>**`if`**</font>.
Portanto, todas as idades menores que `18` ser√£o executadas pelo <font color='orange'>**`else`**</font>.

#üßÆ Estrutura b√°sica: Elif

O exemplo que foi usado tinha apenas 2 op√ß√µes:


*   Maior de idade
*   Menor de idade

Acontece que nem sempre √© assim. Podemos usar um novo exemplo, como classificar pela idade em:

*   Crian√ßa
*   Adolescente
*   Adulto

Perceba que dessa forma, existe mais de 1 possibilidade, ent√£o s√≥ o <font color='orange'>**`if`**</font> e o <font color='orange'>**`else`**</font> n√£o seriam suficientes. Para isso, existe o <font color='orange'>**`elif`**</font> (precisamos admitir que eles mandaram bem no nome).

Essa ferramenta nos permite impor m√∫ltiplas condi√ß√µes.


```python
idade = 10
if idade < 12:
  print("Crian√ßa")
elif idade < 18:
  print ("Adolescente")
else:
  print ("Adulto")
```

***
<font color="grey">
<b>Perceba:</b>
  
  O <font color='orange'>**`else`**</font> s√≥ acontece quando **nenhuma** das outras condi√ß√µes √© atendida.

***

√â importante ter em mente que as condi√ß√µes impostas nos <font color='orange'>**`if`**</font>s, nos <font color='orange'>**`elif`**</font>s e no<font color='orange'>** `else` **</font> s√£o excludentes entre si! Ou seja, se o programa entender que o primeiro <font color='orange'>**`if`**</font> √© verdadeiro, vai devolver o que estiver identado logo depois do <font color='orange'>**`if`**</font> e ignorar as outras op√ß√µes.

####Exemplo 1

```python
x = 30
if (x % 2) == 0:
    print(f'{x} √© par')
elif x > 20:
    print(f'{x} √© maior que 20')
```

Observe que, nesse caso, as condi√ß√µes de <font color='orange'> **`if`**</font> e <font color='orange'>**`elif`**</font>  que estipulamos, na verdade,  n√£o s√£o excludentes. `30` √©, de fato, tanto par quanto maior que `20`, mas o programa n√£o entende isso sozinho. √â necess√°rio que voc√™ mostre isso usando<font color='orange'> **`if`</font> para ambas as condi√ß√µes**, assim ele vai entender que s√£o duas condi√ß√µes independentes.

####Exemplo 2

```python
x = 30
if (x % 2) == 0:
    print(f'{x} √© par')
if x > 20:
    print(f'{x} √© maior que 20')
```

Outro diferencial do Python em quest√£o de simplicidade √© que voc√™ pode montar **todo esse esquema de condi√ß√µes em uma √∫nica senten√ßa**. Fica quase uma frase! Para quem nunca programou com outras linguagens antes n√£o aparece o tamanho da evolu√ß√£o que isso √©, mas confia! O neg√≥cio √© simples mesmo.

```python
nota = 4.5
resultado = 'Aprovado' if nota > 5.75 else 'N√£o aprovado'
print(resultado)
```

```python
nota = 4.5
if nota > 5.75:
    resultado = 'Aprovado'
else:
    resultado = 'N√£o aprovado'
print(resultado)
```
