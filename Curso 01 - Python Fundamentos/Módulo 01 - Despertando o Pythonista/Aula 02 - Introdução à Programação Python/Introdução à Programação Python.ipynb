#🧠 A História da Linguagem Python

## 👤 O que é uma linguagem de programação

aeea21312.png

* **Ser Humano:** Tudo começa com nós, humanos, que temos uma ideia a resolver — um problema, uma tarefa, um processo a automatizar.
* **Ideia**: A ideia é o ponto de partida abstrato. Exemplo: “Quero um programa que calcule a média de três notas.”
* **Raciocínio Lógico**: Para transformar a ideia em algo que o computador possa entender, precisamos organizar o pensamento de forma lógica. Aqui entram: Condições (if/else),Repetições (loops), Etapas sequenciais (passo a passo)
* **Lógica de Programação**: A lógica de programação é a estrutura mental que permite converter raciocínio em instruções organizadas. É o alicerce da programação. É aqui que entram os paradigmas, como veremos.
* **Algoritmo**: Um algoritmo é a descrição ordenada, clara e finita dos passos necessários para resolver um problema.
* **Linguagem de Programação**: Agora, o algoritmo é traduzido para uma linguagem que o computador consiga interpretar — como Python, JavaScript, C++, etc. Essas linguagens têm: Sintaxe (como escrever), Semântica (como interpretar)e Regras próprias.
* **Programa de Computador**: O código escrito na linguagem de programação forma o programa, ou seja, o conjunto de instruções que executam uma função específica.
* **Código Fonte**: O que escrevemos é o código fonte: um arquivo legível por humanos, mas incompreensível pela máquina em sua forma bruta.
* **Linguagem de Máquina**: O código fonte precisa ser convertido para linguagem de máquina (binária: 0s e 1s) para que o computador possa executá-lo. Essa conversão pode ocorrer de duas formas: Por um compilador (ex: C) ou Por um interpretador (ex: Python).
* **Máquina**: Por fim, a máquina executa o programa — obedecendo às instruções binárias resultantes — e realiza a tarefa solicitada (exibir a média, somar dados, gerar um gráfico, etc).




> Compreender esse fluxo ajuda a perceber que programar é mais do que escrever código — é uma forma estruturada de transformar pensamento em ação concreta via tecnologia.



OBS: A diferença entre Compilação e Interpretação é uma das diferenças mais fundamentais entre linguagens de programação: como o código fonte é transformado em algo que a máquina pode executar. Isso envolve dois processos: compilação e interpretação.

| Característica            | Compilador                  | Interpretador              |
|---------------------------|-----------------------------|----------------------------|
| Forma de execução         | Compila tudo antes          | Executa linha por linha    |
| Gera executável?          | Sim                         | Não                        |
| Velocidade de execução    | Alta                        | Menor                      |
| Flexibilidade para testes | Menor                       | Maior                      |
| Necessidade do código     | Não após compilado          | Sempre necessário          |
| Exemplos                  | C, C++, Go                  | Python, JavaScript, R      |

## 🔎 Tipos de Linguagens e Onde o Python se situa

eaae112.png

As linguagens de programação podem ser classificadas conforme vários critérios.


#### 🟩 **Paradigmas de Programação**

Um paradigma é um estilo ou modelo de organização do raciocínio computacional. Eles não são mutuamente excludentes.



> Muitas linguagens (como o Python) suportam múltiplos paradigmas.



* Imperativo: Nesse caso, você diz como o programa deve fazer algo, passo a passo. Características: Baseado em comandos e mudanças de estado. Usa estruturas como if, for, while, variáveis mutáveis. É como uma receita de bolo

```python
# Python é uma linguagem imperativa por padrão.

x = 10
x = x + 5
print(x)
```

* Procedural (subparadigma do imperativo): Nesse caso, Organiza o código em funções/procedimentos reutilizáveis. Características: Ênfase na modularização, Evita repetição de código, Separação clara entre dados e procedimentos

```python
# Python dá suporte completo à programação procedural.

def saudacao(nome):
    print(f"Olá, {nome}!")

saudacao("Gregório")
```

* Orientado a Objetos (OO): Modela o mundo em objetos com atributos e comportamentos. Características: Dados e funções encapsulados em classes, Reutilização com herança, polimorfismo, Ideal para modelar entidades complexas (pacientes, pedidos, sensores)

```python
# Python é orientado a objetos desde sua base: tudo é objeto (strings, funções, números, etc).

class Paciente: # Essa linha cria uma classe chamada Paciente. Classe é uma fábrica de objetos: um molde que define como os objetos do tipo Paciente vão se comportar.

    def __init__(self, nome): # É uma função especial do Python que inicializa o objeto.
        self.nome = nome # É uma referência ao próprio objeto que está sendo criado (como o "eu" do objeto). Nome é um parâmetro externo recebido no momento da criação do paciente.

    def cumprimentar(self): # Esse é um método da classe — uma função que pertence à classe Paciente.
        print(f"Olá, sou {self.nome}") # Quando você chama esse método, ele acessa os dados do próprio objeto (como self.nome) e executa uma ação.


p = Paciente("João") # Essa linha cria um novo objeto do tipo Paciente, passando "João" como argumento.
p.cumprimentar() # Aqui, chamamos o método cumprimentar do objeto p.
```

* Funcional: Foca em funções puras, imutabilidade e ausência de efeitos colaterais. Características: Usa map, filter, reduce, recursão. Funções como valores (funções que recebem e retornam funções). Menor mutação de estado → código mais previsível



```python
# Python não é funcional puro, mas tem suporte funcional suficiente para muitos usos.

dobro = lambda x: x * 2
print(list(map(dobro, [1, 2, 3, 4])))
```

* Paradigma Declarativo: No paradigma declarativo, você descreve o que deseja obter, não como isso deve ser feito. O foco está no resultado, não nos passos.

```python
# SQL é declarativo. Você só diz o que quer — os dados. Não especifica como percorrer os registros.

'SELECT nome FROM pacientes WHERE idade > 60'
```

```python
# Em Python você precisa, normalmente: Explicitar o controle de fluxo. Mas é possível simular o estilo declarativo, como na biblioteca Pandas:

"df[df['idade'] > 60]['nome']"
```

#### 🟩 **Estrutura de Tipos**

A estrutura de tipos define como a linguagem lida com os tipos de dados (como int, str, bool, etc).

| Tipo                  | Significado                                                                                     |
| --------------------- | ----------------------------------------------------------------------------------------------- |
| **Estática**          | O tipo é definido **antes da execução**, geralmente no código fonte. Ex: C, Java.               |
| **Dinâmica**          | O tipo é definido **em tempo de execução**, baseado no valor atribuído. Ex: Python, JavaScript. |
| **Fortemente tipada** | Impede operações implícitas entre tipos incompatíveis. Ex: Python, Java.                        |
| **Fracamente tipada** | Permite misturas arriscadas, fazendo conversões automáticas. Ex: JavaScript, C.                 |


Python tem tipagem Dinâmica (não precisa declarar tipo). E é Fortemente tipada (não converte automaticamente).

```python
x = 10       # tipo int
x = "dez"    # agora é string

print(x + 1) # Erro! Não mistura string com int automaticamente
```

#### 🟩 **Grau de Abstração**

Esse eixo classifica as linguagens com base na distância entre o código e a máquina.

| Grau            | Características                                           | Exemplo                  |
| --------------- | --------------------------------------------------------- | ------------------------ |
| **Baixo nível** | Próxima da linguagem de máquina                           | Assembly, C              |
| **Médio nível** | Mistura instruções mais abstratas com controle de memória | C++, Rust                |
| **Alto nível**  | Próxima da linguagem humana, abstrai a máquina            | Python, JavaScript, Ruby |


Python é uma linguagem de alto nível:
* Abstrai totalmente a memória
* Não usa ponteiros
* Foca em clareza e legibilidade
* Permite escrever menos código para fazer mais

#### 🟩 **Geração da Linguagem**

A classificação por gerações é histórica e representa a evolução da linguagem de máquina para linguagens mais expressivas.

| Geração    | Características                                        | Exemplo             |
| ---------- | ------------------------------------------------------ | ------------------- |
| 1ª Geração | Código de máquina binário (0 e 1)                      | Instruções nativas  |
| 2ª Geração | Assembly, usa mnemônicos para instruções               | MOV, ADD            |
| 3ª Geração | Linguagens estruturadas e imperativas                  | C, Java             |
| 4ª Geração | Linguagens voltadas à produtividade, mais declarativas | SQL, Python, MATLAB |
| 5ª Geração | Foco em IA e lógica (baseadas em inferência)           | Prolog, Mercury     |


Python é tradicionalmente considerado uma linguagem de 3ª geração, mas por sua:
* Alta expressividade
* Sintaxe próxima da linguagem natural
* Foco em produtividade

É frequentemente considerada também uma linguagem de 4ª geração moderna.

No final dos anos 1980, na Holanda 🇳🇱, Guido estava trabalhando no centro de pesquisa CWI e queria criar uma linguagem de programação que fosse:
* Mais simples de escrever do que C ou Java
* Fácil de entender (até por quem nunca programou)
* Aberta e acessível a todos

Ele lançou a primeira versão oficial do Python em 1991.

## 🧘 Princípios: O Zen do Python (The Zen of Python)

São 19 princípios escritos por Tim Peters, um dos contribuidores da linguagem

* Beautiful is better than ugly (Belo é melhor que feio.)
* Explicit is better than implicit (Explícito é melhor que implícito.)
* Simple is better than complex (Simples é melhor que complexo.)
* Complex is better than complicated (Complexo é melhor que complicado.)
* Flat is better than nested (Plano é melhor que aninhado.)
* Sparse is better than dense (Espaçado é melhor que denso.)
* Readability counts (Legibilidade conta.)
* Special cases aren't special enough to break the rules. (Casos especiais não são especiais o suficiente para quebrar as regras.)
* Although practicality beats purity (Embora a praticidade vença a pureza.)Errors should never pass silently (Erros nunca devem passar silenciosamente.) Unless explicitly silenced (A não ser que sejam explicitamente silenciados.)
* In the face of ambiguity, refuse the temptation to guess (Diante da ambiguidade, recuse a tentação de adivinhar.)
* There should be one-- and preferably only one --obvious way to do it (Deve haver uma — e preferencialmente apenas uma — maneira óbvia de fazer algo.)
* Although that way may not be obvious at first unless you're Dutch.
(Embora essa maneira possa não ser óbvia de início — a menos que você seja holandês.)
* Now is better than never (Agora é melhor do que nunca.)
* Although never is often better than right now.
(Embora nunca seja frequentemente melhor do que agora.) If the implementation is hard to explain, it's a bad idea (Se a implementação for difícil de explicar, é uma má ideia.)
* If the implementation is easy to explain, it may be a good idea (Se a implementação for fácil de explicar, pode ser uma boa ideia.)
* Namespaces are one honking great idea -- let's do more of those!
(Namespaces são uma grande ideia — vamos usá-los mais!)
