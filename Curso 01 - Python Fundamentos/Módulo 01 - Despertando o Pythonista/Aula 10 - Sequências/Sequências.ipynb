# üëâ Sequ√™ncias

Listas, strings e tuplas s√£o todas exemplos de <font color = "orange"><b>sequ√™ncias </font></b> - uma s√©rie de **elementos ordenados** (ou de caracteres, no caso das strings)

```python
minha_lista = ['a', 'b', 'c']
minha_tupla = ('a', 'b', 'c')
minha_string = 'abc'
```

###Indexa√ß√£o

<font color = "orange"><b>`seq[n]`</font></b> vai retornar o elemento (ou caracter) na n-√©sima posi√ß√£o da sequ√™ncia em quest√£o.

```python
if minha_lista[1] == minha_tupla[1] == minha_string[1]:
    print(minha_lista[1], minha_tupla[1], minha_string[1])
```

###Comprimento

<font color = "orange"><b>`len(seq)`</font></b>, como j√° vimos antes, retorna o comprimento da sequ√™ncia, ou seja, a quantidade de elementos ou caracteres.

```python
if len(minha_lista) == len(minha_tupla) == len(minha_string):
    print(len(minha_lista), len(minha_tupla), len(minha_string))
```

###Segmenta√ß√£o

Assim como j√° fizemos com as listas, podemos segmentar qualquer sequ√™nci

```python
print (minha_tupla[1:]) # Esse [1:] significa que a tupla ser√° segmentada
                        #a partir do elemento de posi√ß√£o 1
                        #e, como n√£o definimos uma posi√ß√£o depois do :
                        #ela vai at√© o fim
```

###Operadores

<font color = "orange"><b>`in`</font></b> e <font color = "orange"><b>`not in`</font></b> s√£o operadores. Assim, voc√™ pode checar se algum elemento √© ou n√£o parte de uma sequ√™ncia. O interpretador vai entender como uma **pergunta** e vai te responder com <font color = "orange"><b>`True`</font></b>  ou <font color = "orange"><b>`False`</font></b>.

***
Exemplo: <font color = "orange"><b>`x`</font></b> in <font color = "orange"><b>`seq`</font></b> == <font color = "orange"><b>`True`</font></b> ‚áî algum item da sequ√™ncia √© igual a x
***

Para strings, voc√™ pode checar se uma substring faz parte da string em quest√£o.

```python
'd' in minha_lista
```

```python
'b' in minha_tupla and 'b' in minha_lista and 'b' in minha_string
```

```python
'odu√ß' in 'Engenharia de produ√ß√£o'
```

###Concatena√ß√£o

Podemos unir objetos do mesmo tipo de maneira bem f√°cil usando o operador de adi√ß√£o <font color = "orange"><b>+</b></font>.

Ou duplic√°-lo quantas vezes quisermos utilizando o operador de multipli√ß√£o <font color = "orange"><b>*</b></font>.

```python
uma_tupla = (1, 2, 3)
uma_tupla
```

```python
uma_tupla + (5, 6, 7) #Lembra que tem que ser um objeto do mesmo tipo (tupla + tupla)
```

```python
uma_tupla
```

```python
"Yay! " * 5
```

Quando voc√™ faz list('de'), o Python quebra a string em uma lista de caracteres. O mesmo acontece com tuple('de'), que vira ('d', 'e').

üîß Isso acontece porque strings s√£o iter√°veis em Python, e list() ou tuple() percorrer√° cada caractere da string individualmente.

```python
print(minha_lista + list('de'),
      minha_tupla + tuple('de'),
      minha_string + 'de')
```
