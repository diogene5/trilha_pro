#### `Series`

A estrutura de dados `Series` √© uma matriz unidimensional, ou seja, uma lista de valores. Ela armazena qualquer tipo de dado (√© simples assim mesmo, n√£o tem mist√©rio). Vamos criar uma `Series` com as notas das p1:

```python
notas.values  #Retorna apenas os valores da Series
```

```python
notas.index  #Retorna o index da Series
```

`DataFrame` √© uma estrutura de dados bidimensional, podendo ter v√°rias colunas, com diferentes tipos de valores, como uma planilha. Vamos criar um `DataFrame`:

  5. O espa√ßo que o `DataFrame` utiliza na mem√≥ria.

Existem casos em que os n√∫meros n√£o suportam ser analisados dessa forma, por exemplo, uma coluna que contenha o n√∫mero da matr√≠cula dos alunos; j√° que s√£o n√∫meros, o Pandas realiza esse m√©todo at√© nela. ### ‚úÖ Selecionando dados

√Äs vezes voc√™ precisa de um dado espec√≠fico para trabalhar. Por exemplo, quais mat√©rias voc√™ tirou notas abaixo da m√©dia?

Com o tempo, voc√™ vai perceber que saber selecionar algum dado, ou um tipo de dado, √© muito importante. Por isso, vamos aprender algumas maneiras de selecionar dados.

* Localiza√ß√£o pelo **`index`**: selecionando valores de uma linha, coluna:

> * <font color = "orange">**`.loc[]` </font>**: ele localiza atrav√©s do **nome** da linha

```python
dados
```

```python
dados.loc['Medicina Geral de Adultos']  #Localizamos todas as informa√ß√µes a respeito da linha com index 'Estat√≠stica'
```

Podemos localizar informa√ß√µes de mais de uma linha:

```python
dados.loc[['Medicina Geral de Adultos', 'Medicina Geral de Crian√ßas']]
```

> * <font color = "orange">**`.iloc[]` </font>**: ele localiza atrav√©s da **posi√ß√£o** (representada por um n√∫mero inteiro) da linha no DataFrame

```python
dados.iloc[0]  #Localizamos as informa√ß√µes da primeira linha do DataFrame
```

```python
dados.iloc[-1]  #Localizamos as informa√ß√µes da √∫ltima linha
```

Podemos incluir a localiza√ß√£o de colunas tamb√©m:

```python
dados.iloc[0:2, 0] #Localizamos as informa√ß√µes da primeira coluna com a primeira e segunda linha
```

* Localiza√ß√£o por crit√©rios **boleanos**: selecionando dados atrav√©s de condi√ß√µes

```python
#Selecionar apenas as linhas que tenham valores menores do que '5.75' na coluna 'P1':

dados[dados['P1'] < 5.75]
```

 > Lembra em Python, quando quer√≠amos associar **m√∫ltiplas condi√ß√µes**, e utiliz√°vamos operadores como o `and`, `or` e `not`?! Aqui √© quase a mesma coisa, s√≥ mudam os nomes desses operadores.

 > * <font color = "orange">**`&` </font>**: Assume a fun√ß√£o do <font color = "orange">**`and` </font>**
 > * <font color = "orange">**`|` </font>**: Assume a fun√ß√£o do <font color = "orange">**`or` </font>**
 > * <font color = "orange">**`~` </font>**: Assume a fun√ß√£o do <font color = "orange">**`not`</font>**

```python
#Selecionar as linhas que tenham valores maiores do que '8' nas colunas 'P1' e 'P2':

dados[(dados['P1'] > 8) & (dados['P2'] > 8)]
```
