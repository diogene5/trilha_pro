```python
SHEET_ID = "1YQhDNSt1sOrzq6AKxyYG6wzRAvM-M7l7jJRIHunNB88" # Pegar o id da planilha

spreadsheet = gc.open_by_key(SHEET_ID) # Abre a planilha pelo ID
```

```python
rows = worksheet.get_all_records() #pega todos os registros da aba
```

```python
df_receitas = pd.DataFrame(rows) #cria um dataframe (uma tabela) com as linhas adquiridas
```

```python
#é possível montar uma função que faz tudo de uma só vez e pode ser facilmente reaproveitável

def aquisicao_google_sheets(sheet_id, sheet_name):
    SHEET_ID = sheet_id
    SHEET_NAME = sheet_name
    gc = gspread.authorize(creds)
    spreadsheet = gc.open_by_key(SHEET_ID)
    worksheet = spreadsheet.worksheet(SHEET_NAME)
    rows = worksheet.get_all_records()
    df = pd.DataFrame(rows)
    return df
```

```python
df_receitas = aquisicao_google_sheets("1YQhDNSt1sOrzq6AKxyYG6wzRAvM-M7l7jJRIHunNB88", "Receitas")
```
