## üß± Passos para Criar um Gr√°fico

1. Criar os dados
2. Criar a figura e os eixos com `plt.subplots()`
3. Tra√ßar os dados com `ax.plot()` ou equivalente
4. Adicionar personaliza√ß√£o como t√≠tulos
5. Exibir com `plt.show()`

```python
import pandas as pd
```

```python
# Criar os dados

df = pd.DataFrame({
    "X": [10, 15, 20, 25, 30, 35, 40],
    "Y": [20, 24, 28, 32, 36, 40, 44]
})
```

```python
df
```

```python
# Criar a figura e os eixos com `plt.subplots()`

fig, ax = plt.subplots()
plt.show()
```

```python
# Tra√ßar os dados com `ax.plot()` ou equivalente

fig,ax = plt.subplots()
ax.plot(df["X"], df["Y"])
plt.show()
```

```python
# Adicionar personaliza√ß√£o como t√≠tulos

fig, ax = plt.subplots()
ax.plot(df["X"], df["Y"])
ax.set_title('Gr√°fico de Linha')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
plt.show()
```

## üìä Tipos de gr√°fico mais comuns

### üìä Gr√°fico de Linha (plot)

Mostra a rela√ß√£o entre vari√°veis cont√≠nuas ao longo de um eixo (normalmente o tempo).

```python
fig, ax = plt.subplots()

plt.plot(df["X"], df["Y"], marker='o', color='blue', label='Linha')
plt.title("Gr√°fico de Linha")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

plt.show()
```

### üìä Gr√°fico de Dispers√£o (scatter)

Mostra a rela√ß√£o entre duas vari√°veis num√©ricas com pontos independentes.

```python
fig, ax = plt.subplots()

plt.scatter(df["X"], df["Y"], color='red', s=80)
plt.title("Gr√°fico de Dispers√£o")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### üìä Gr√°fico de Barras (bar)

Compara valores entre categorias.

```python
plt.bar(df["X"], df["Y"], color='orange', edgecolor='black')
plt.title("Gr√°fico de Barras")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.show()
```

### üìä Gr√°fico de Barras Horizontais (barh)

Igual ao de barras, mas com orienta√ß√£o horizontal.

```python
plt.barh(df["X"], df["Y"], color='green')
plt.title("Gr√°fico de Barras Horizontais")
plt.xlabel("Valores")
plt.ylabel("Categorias")
plt.show()
```

### üìä Gr√°fico de Pizza

```python
labels = ['A', 'B', 'C']
sizes = [30, 45, 25]
colors = ['#ff9999','#66b3ff','#99ff99']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # deixa circular
plt.show()
```

## üé® Personaliza√ß√£o detalhada do Gr√°fico

### üìè T√≠tulo Principal

* **Fun√ß√£o**: Define o t√≠tulo principal do gr√°fico
* **C√≥digo:** set_title()
* **Principais par√¢metros**:
  * label (posicional) ‚Üí texto do t√≠tulo.
  * fontsize ‚Üí tamanho da fonte.
  * fontweight ‚Üí peso da fonte ('normal', 'bold', etc.).
  * loc ‚Üí alinhamento do t√≠tulo ('center', 'left', 'right').
  * pad ‚Üí espa√ßamento do t√≠tulo.

```python
# Exemplo de implementa√ß√£o

#Cria√ß√£o da figura
fig, ax = plt.subplots()

#Plotagem dos dados
ax.plot(df["X"], df["Y"])

#Personaliza√ß√£o do t√≠tulo
ax.set_title("T√≠tulo Principal",
             fontsize=14,
             fontweight='bold',
             loc='center',
             pad=30)

#Plotagem
plt.show()
```

### üìè T√≠tulo do Eixo X

* **Fun√ß√£o**: Define o r√≥tulo do eixo X (horizontal)
* **C√≥digo**: set_xlabel()
* **Principais par√¢metros**:
  * xlabel (posicional) ‚Üí texto do r√≥tulo.
  * fontsize ‚Üí tamanho da fonte.
  * labelpad ‚Üí espa√ßamento entre o texto e o eixo.

```python
# Exemplo de implementa√ß√£o

fig, ax = plt.subplots()
ax.plot(df["X"], df["Y"])
ax.set_title("T√≠tulo Principal", fontsize=14, fontweight='bold', loc='center', pad=30)
ax.set_xlabel("Eixo X", fontsize=12, pad=10)
plt.show()
```

### üìè T√≠tulo do Eixo Y

* **Fun√ß√£o**: Define o r√≥tulo do eixo Y (vertical).
* **C√≥digo:** set_ylabel()
* **Principais par√¢metros**:
  * ylabel (posicional) ‚Üí texto do r√≥tulo.
  * fontsize ‚Üí tamanho da fonte.
  * labelpad ‚Üí espa√ßamento entre o texto e o eixo.

```python
# Exemplo de implementa√ß√£o

fig, ax = plt.subplots()
ax.plot(df["X"], df["Y"])
ax.set_title("T√≠tulo Principal", fontsize=14, fontweight='bold', loc='center', pad=30)
ax.set_xlabel("Eixo X", fontsize=12, labelpad=10)
ax.set_ylabel("Eixo Y", fontsize=12, labelpad=10)
plt.show()
```

### üìã Legenda

* **Fun√ß√£o**: Exibe uma legenda para os elementos gr√°ficos (linhas, pontos, barras).
* **C√≥digo:** O texto exibido na legenda vem do par√¢metro label usado na fun√ß√£o de plotagem (ax.plot(), ax.scatter(), etc.).
* **Principais par√¢metros**:
  * loc ‚Üí posi√ß√£o da legenda ('upper right', 'upper left', 'lower left', 'lower right', 'center', etc.).
  * bbox_to_anchor ‚Üí coordenadas para posicionamento manual. Voc√™ passa coordenadas relativas (entre 0 e 1) para dizer onde esse ponto vai estar.
  * fontsize ‚Üí tamanho do texto.
  * title ‚Üí t√≠tulo da legenda.
  * frameon ‚Üí exibir ou n√£o a caixa da legenda (True ou False).
  * facecolor, edgecolor, framealpha ‚Üí personaliza√ß√£o do fundo, borda e transpar√™ncia

```python
# Exemplo de implementa√ß√£o

fig, ax = plt.subplots()
ax.plot(df["X"], df["Y"], label='Dados')
ax.set_title("T√≠tulo Principal", fontsize=14, fontweight='bold', loc='center', pad=30)
ax.set_xlabel("Eixo X", fontsize=12, labelpad=10)
ax.set_ylabel("Eixo Y", fontsize=12, labelpad=10)
ax.legend(loc='lower center', bbox_to_anchor=(0.90, 1), fontsize=10, title="Legenda")
plt.show()
```

### üìå Limites dos eixos

* **Fun√ß√£o**: Define manualmente os limites do eixo X ou Y. Para focar em uma parte espec√≠fica dos dados e para manter escala consistente entre m√∫ltiplos gr√°ficos.
* **C√≥digo:** set_xlim() e set_ylim()
* **Principais par√¢metros**:
  * (min, max) ‚Üí valores num√©ricos para in√≠cio e fim do eixo.

```python
# Exemplo de implementa√ß√£o

fig, ax = plt.subplots()
ax.plot(df["X"], df["Y"], label='Dados')
ax.set_title("T√≠tulo Principal", fontsize=14, fontweight='bold', loc='center', pad=30)
ax.set_xlabel("Eixo X", fontsize=12, labelpad=10)
ax.set_ylabel("Eixo Y", fontsize=12, labelpad=10)
ax.legend(loc='lower center', bbox_to_anchor=(0.90, 1), fontsize=10, title="Legenda")
ax.set_xlim(10, 40)  # Eixo X de 10 a 40
ax.set_ylim(20, 45)  # Eixo Y de 20 a 45
plt.show()
```

### üìå Marca√ß√µes dos eixos

* **Fun√ß√£o**: Personaliza a apar√™ncia e o comportamento dos ticks (marcas nos eixos) e seus r√≥tulos.
* **C√≥digo:** tick_params()
* **Principais par√¢metros**:
  * axis ‚Üí 'x', 'y' ou 'both'.
  * rotation ‚Üí rota√ß√£o do texto do r√≥tulo.
  * labelsize ‚Üí tamanho da fonte.
  * length ‚Üí comprimento da marca.
  * width ‚Üí espessura da marca.
  * colors ‚Üí cor dos r√≥tulos e ticks.

```python
# Exemplo de implementa√ß√£o

fig, ax = plt.subplots()
ax.plot(df["X"], df["Y"], label='Dados')
ax.set_title("T√≠tulo Principal", fontsize=14, fontweight='bold', loc='center', pad=30)
ax.set_xlabel("Eixo X", fontsize=12, labelpad=10)
ax.set_ylabel("Eixo Y", fontsize=12, labelpad=10)
ax.legend(loc='lower center', bbox_to_anchor=(0.90, 1), fontsize=10, title="Legenda")
ax.set_xlim(10, 40)  # Eixo X de 10 a 40
ax.set_ylim(20, 45)  # Eixo Y de 20 a 45
ax.tick_params(axis='both', rotation=45, labelsize=12)  # Rotaciona r√≥tulos do eixo X em 45¬∞
plt.show()
```

### üìå Escalas dos eixos

* **Fun√ß√£o**: Por padr√£o, os eixos de um gr√°fico no Matplotlib usam escala linear, ou seja, os valores aumentam de forma uniforme.
* **C√≥digo:** set_xscale() e set_yscale()
* **Principais par√¢metros**: tipo pode ser:
  * 'linear' ‚Üí padr√£o, espa√ßamento uniforme.
  * 'log' ‚Üí escala logar√≠tmica (base 10 por padr√£o).
  * 'symlog' ‚Üí logar√≠tmica sim√©trica (aceita valores negativos pr√≥ximos de zero).
  * 'logit' ‚Üí escala log√≠stica (para propor√ß√µes entre 0 e 1).

```python
# Exemplo de implementa√ß√£o

fig, ax = plt.subplots()
ax.plot(df["X"], df["Y"], label='Dados')
ax.set_title("T√≠tulo Principal", fontsize=14, fontweight='bold', loc='center', pad=30)
ax.set_xlabel("Eixo X", fontsize=12, labelpad=10)
ax.set_ylabel("Eixo Y", fontsize=12, labelpad=10)
ax.legend(loc='lower center', bbox_to_anchor=(0.90, 1), fontsize=10, title="Legenda")
ax.set_xscale('log')
ax.set_yscale('log')
ax.tick_params(axis='both', rotation=45, labelsize=12)  # Rotaciona r√≥tulos do eixo X em 45¬∞
plt.show()
```

### üéØ Tamanho da figura

* **C√≥digo:** plt.figure(figsize=(largura, altura)) ou diretamente em plt.subplots(figsize=(largura, altura))

### üîß Subgr√°ficos

* **Fun√ß√£o**:  "sub-plotagem", ou seja, um gr√°fico dentro de uma grade de gr√°ficos na mesma figura. Muito √∫til para comparar diferentes visualiza√ß√µes lado a lado.
* **C√≥digo:** fig, axs = plt.subplots(nrows, ncols, figsize=(largura, altura))
* **Principais par√¢metros**:
  * nrows ‚Üí n√∫mero de linhas.
  * ncols ‚Üí n√∫mero de colunas.
  * axs ‚Üí objeto ou lista de objetos dos eixos criados.

```python
# Exemplo de implementa√ß√£o

fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 linha, 2 colunas

# Primeiro gr√°fico
axs[0].plot(df["X"], df["Y"], marker='o')
axs[0].set_title("Gr√°fico 1")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")

# Segundo gr√°fico
axs[1].bar(df["X"], df["Y"], color='orange')
axs[1].set_title("Gr√°fico 2")
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")

plt.tight_layout() # ajustando espa√ßos entre subplots
plt.show()
```

### üìä Linhas de grade

* **Fun√ß√£o**: Adiciona uma grade ao gr√°fico para facilitar a leitura dos valores.
* **C√≥digo:** grid()
* **Principais par√¢metros**:
  * True ou False ‚Üí ativa/desativa a grade.
  * axis ‚Üí 'x', 'y' ou 'both'.
  * linestyle ‚Üí estilo da linha ('--', '-', ':').
  * color ‚Üí cor da grade.
  * alpha ‚Üí transpar√™ncia.

```python
# Exemplo de implementa√ß√£o

fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 linha, 2 colunas

# Primeiro gr√°fico
axs[0].plot(df["X"], df["Y"], marker='o')
axs[0].set_title("Gr√°fico 1")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")
axs[0].grid(True, linestyle='--', alpha=0.5)

# Segundo gr√°fico
axs[1].bar(df["X"], df["Y"], color='orange')
axs[1].set_title("Gr√°fico 2")
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")
axs[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout() # ajustando espa√ßos entre subplots
plt.show()
```

### üìùAnota√ß√µes

* **Fun√ß√£o**: Adiciona texto e setas para destacar pontos espec√≠ficos no gr√°fico.
* **C√≥digo:** annotate()
* **Principais par√¢metros**:
  * text ‚Üí texto a ser exibido.
  * xy ‚Üí coordenada do ponto destacado.
  * xytext ‚Üí coordenada onde o texto ficar√°.
  * arrowprops ‚Üí dicion√°rio com propriedades da seta (cor, estilo).

```python
# Exemplo de implementa√ß√£o

fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 linha, 2 colunas

# Primeiro gr√°fico
axs[0].plot(df["X"], df["Y"], marker='o')
axs[0].set_title("Gr√°fico 1")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")
axs[0].grid(True, linestyle='--', alpha=0.5)

# Anota√ß√£o com seta vindo de cima
axs[0].annotate(
    "Importante", # Texto exibido no gr√°fico
    xy=(30, 36),  # Coordenadas do ponto alvo
    xytext=(30, 42), # Coordenadas onde o texto ser√° colocado
    ha='center', # Alinhamento horizontal do texto
    arrowprops=dict(  # Propriedades da seta
        facecolor='black',  # Cor de preenchimento da seta
        arrowstyle='->',    # Estilo da seta (seta simples)
        lw=1.5    # Espessura da linha da seta
    )
)

# Segundo gr√°fico
axs[1].bar(df["X"], df["Y"], color='orange')
axs[1].set_title("Gr√°fico 2")
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")
axs[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout() # ajustando espa√ßos entre subplots
plt.show()
```

### üìùLinha de refer√™ncia

* **Fun√ß√£o**: Desenha linhas de refer√™ncia horizontal (axhline) ou vertical (axvline).
* **C√≥digo:** axhline() / axvline()
* **Principais par√¢metros**:
  * y ou x ‚Üí posi√ß√£o da linha.
  * color ‚Üí cor da linha.
  * linestyle ‚Üí estilo da linha.
  * lw ‚Üí espessura.

```python
# Exemplo de implementa√ß√£o

fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 linha, 2 colunas

# Primeiro gr√°fico
axs[0].plot(df["X"], df["Y"], marker='o')
axs[0].set_title("Gr√°fico 1")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")
axs[0].grid(True, linestyle='--', alpha=0.5)

# Anota√ß√£o com seta vindo de cima
axs[0].annotate(
    "Importante", # Texto exibido no gr√°fico
    xy=(30, 36),  # Coordenadas do ponto alvo
    xytext=(30, 42), # Coordenadas onde o texto ser√° colocado
    ha='center', # Alinhamento horizontal do texto. Pode ser: 'center' ‚Üí centraliza, 'left' ‚Üí alinha √† esquerda, 'right' ‚Üí alinha √† direita.
    arrowprops=dict(  # Propriedades da seta
        facecolor='black',  # Cor de preenchimento da seta
        arrowstyle='->',    # Estilo da seta. Poder ser: aqui √© seta simples, mas pode ser '-|>', '<->', etc.
        lw=1.5,    # Espessura da linha da seta
        linestyle='--'  # Estilo da linha da seta. Pode ser: '-', '--', '-.', ':'
    )
)

# Linhas de refer√™ncia
axs[0].axhline(y=30, color='red', linestyle=':')
axs[0].axvline(x=25, color='blue', linestyle='--')

# Segundo gr√°fico
axs[1].bar(df["X"], df["Y"], color='orange')
axs[1].set_title("Gr√°fico 2")
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")
axs[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout() # ajustando espa√ßos entre subplots
plt.show()
```

### üìù Destaque de √°rea

* **Fun√ß√£o**: Destaca uma √°rea horizontal (axhspan) ou vertical (axvspan) com um ret√¢ngulo sombreado.
* **C√≥digo:** axhspan() / axvspan()
* **Principais par√¢metros**:
  * In√≠cio e fim (no eixo correspondente).
  * color ‚Üí cor de preenchimento.
  * alpha ‚Üí transpar√™ncia.

```python
# Exemplo de implementa√ß√£o

fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 linha, 2 colunas

# Primeiro gr√°fico
axs[0].plot(df["X"], df["Y"], marker='o')
axs[0].set_title("Gr√°fico 1")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")
axs[0].grid(True, linestyle='--', alpha=0.5)

# Anota√ß√£o com seta vindo de cima
axs[0].annotate(
    "Importante", # Texto exibido no gr√°fico
    xy=(30, 36),  # Coordenadas do ponto alvo
    xytext=(30, 42), # Coordenadas onde o texto ser√° colocado
    ha='center', # Alinhamento horizontal do texto. Pode ser: 'center' ‚Üí centraliza, 'left' ‚Üí alinha √† esquerda, 'right' ‚Üí alinha √† direita.
    arrowprops=dict(  # Propriedades da seta
        facecolor='black',  # Cor de preenchimento da seta
        arrowstyle='->',    # Estilo da seta. Poder ser: aqui √© seta simples, mas pode ser '-|>', '<->', etc.
        lw=1.5,    # Espessura da linha da seta
        linestyle='--'  # Estilo da linha da seta. Pode ser: '-', '--', '-.', ':'
    )
)

# Linhas de refer√™ncia
axs[0].axhline(y=30, color='red', linestyle=':')
axs[0].axvline(x=25, color='blue', linestyle='--')

# √Åreas sombreadas
axs[0].axvspan(20, 30, color='yellow', alpha=0.3)

# Segundo gr√°fico
axs[1].bar(df["X"], df["Y"], color='orange')
axs[1].set_title("Gr√°fico 2")
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")
axs[1].grid(True, linestyle='--', alpha=0.5)

plt.tight_layout() # ajustando espa√ßos entre subplots
plt.show()
```

### #Ô∏è‚É£ R√≥tulo de valores

* **Fun√ß√£o**: Adiciona valores diretamente sobre os pontos ou barras do gr√°fico para facilitar a interpreta√ß√£o visual.
* **Principais par√¢metros**:
  * padding ‚Üí dist√¢ncia entre o texto e a barra (bar_label).
  * fontsize ‚Üí tamanho da fonte.
  * color ‚Üí cor do texto.
  * ha ‚Üí alinhamento horizontal ('center', 'left', 'right').
  * va ‚Üí alinhamento vertical ('bottom', 'top').
  * xytext ‚Üí deslocamento do texto (em annotate).

* **C√≥digo:**

| Tipo de gr√°fico        | Melhor forma de colocar r√≥tulos |
| ---------------------- | ------------------------------- |
| Barras (`bar`, `barh`) | `ax.bar_label()`                |
| Linha (`plot`)         | `ax.text()` ou `ax.annotate()`  |
| Dispers√£o (`scatter`)  | `ax.text()` ou `ax.annotate()`  |

```python
def plot_exemplo_unico_linha(df):

    """
    Gera um gr√°fico de linha com:
    - T√≠tulos e r√≥tulos de eixos
    - Grid
    - Anota√ß√£o com seta
    - Linhas de refer√™ncia
    - √Årea sombreada
    - R√≥tulos de valores sobre os pontos
    """

    # ========================
    # 1. Cria√ß√£o da figura
    # ========================
    fig, ax = plt.subplots(figsize=(8, 5))

    # ========================
    # 2. Defini√ß√£o do tipo de gr√°fico
    # ========================
    ax.plot(df["X"], df["Y"], marker='o', label="Dados", color='blue')

    # ========================
    # 3. Defini√ß√£o dos t√≠tulos
    # ========================
    ax.set_title("T√≠tulo Principal", fontsize=14, fontweight='bold', loc='center', pad=30)
    ax.set_xlabel("Eixo X", fontsize=12, labelpad=10)
    ax.set_ylabel("Eixo Y", fontsize=12, labelpad=10)

    # ========================
    # 4. Defini√ß√£o da legenda
    # ========================
    ax.legend(loc='lower center', bbox_to_anchor=(0.90, 1), fontsize=10, title="Legenda")

    # ========================
    # 5. Defini√ß√£o do limite dos eixos
    # ========================
    ax.set_xlim(10, 40)
    ax.set_ylim(20, 45)

    # ========================
    # 6. Defini√ß√£o das marca√ß√µes dos eixos
    # ========================
    ax.tick_params(axis='both', rotation=0, labelsize=12)

    # ========================
    # 7. Defini√ß√£o das linhas de grade
    # ========================
    ax.grid(True, linestyle='--', alpha=0.5)

    # ========================
    # 8. Defini√ß√£o de linhas de refer√™ncia
    # ========================
    ax.axhline(y=32, color='red', linestyle=':', linewidth=1)
    ax.axvline(x=25, color='blue', linestyle='--', linewidth=1)

    # ========================
    # 9. Defini√ß√£o de r√≥tulos de valores
    # ========================
    for x_val, y_val in zip(df["X"], df["Y"]):
        ax.text(x_val, y_val + 0.5, f"{y_val}",
                ha='center', va='bottom', fontsize=9, color='black')

    # ========================
    # 10. Defini√ß√£o de anota√ß√µes
    # ========================
    ax.annotate(
        "Importante",
        xy=(30, 36),  # ponto alvo
        xytext=(30, 42),  # posi√ß√£o do texto
        ha='center',
        arrowprops=dict(
            facecolor='black',
            arrowstyle='->',
            lw=1.5,
            linestyle='--'
        )
    )

    # ========================
    # 11. Defini√ß√£o de √°rea de destaque
    # ========================
    ax.axvspan(20, 30, color='yellow', alpha=0.3)

    # ========================
    # 12. Plotagem
    # ========================
    plt.show()
```

## üíº Aplica√ß√£o em Cen√°rio

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
from datetime import datetime, timedelta
```

### Dados sint√©ticos

```python
# -----------------------
# Dados sint√©ticos (exemplo)
# -----------------------

def gerar_dados_pa_com_outliers(
    dias=14,
    seed=42,
    outlier_hours=(10, 17),   # hor√°rios que ter√£o pico (ex.: 10h e 17h)
    outlier_days=2,           # em quantos dias distintos injetar pico
    outlier_factor=2.8,       # intensidade do pico (>=2 costuma ser percept√≠vel)
    usar_hora="atendimento"   # "atendimento" ou "chegada" para preencher coluna 'hora'
):
    """
    Gera dados sint√©ticos de PA com chegada, triagem, atendimento, especialidade,
    profissional e classifica√ß√£o de risco (Manchester), e injeta picos (outliers)
    em dias/horas selecionados.

    Retorna:
        df (DataFrame) com colunas:
            ['atendimento_id','data_hora_chegada','data_hora_triagem',
             'data_hora_atendimento','especialidade','profissional_id',
             'hora','classificacao_risco']
    """
    np.random.seed(seed)
    random.seed(seed)

    # -----------------------
    # Especialidades e plantel
    # -----------------------
    especialidades = ["Cl√≠nica", "Pediatria", "Cirurgia", "Ortopedia"]
    prof_por_espec = {
        "Cl√≠nica":   [f"CLI_{i:02d}" for i in range(1, 16)],
        "Pediatria": [f"PED_{i:02d}" for i in range(1, 11)],
        "Cirurgia":  [f"CIR_{i:02d}" for i in range(1, 9)],
        "Ortopedia": [f"ORT_{i:02d}" for i in range(1, 9)],
    }

    # -----------------------
    # Perfil hor√°rio (24h) ‚Äî picos 10‚Äì11, 16‚Äì18; vale 13‚Äì14; queda √† noite e repique 22h
    # -----------------------
    HOUR_MULT = np.array([
        0.35, 0.40, 0.45, 0.50, 0.55, 0.65, 0.85, 1.10, 1.40, 1.70, 1.95, 1.85,  # 00‚Äì11
        0.90, 0.80, 1.10, 1.35, 1.80, 1.90,                                      # 12‚Äì18
        1.30, 0.80, 0.65, 0.75, 1.20, 0.95                                       # 19‚Äì23
    ])

    # base de atendimentos por prof/hora (antes do perfil)
    BASE_LAMBDA = {"Cl√≠nica": 1.9, "Pediatria": 1.1, "Cirurgia": 0.55, "Ortopedia": 0.65}

    def n_prof_ativos(espec: str, hora: int) -> int:
        base = {"Cl√≠nica": 8, "Pediatria": 5, "Cirurgia": 3, "Ortopedia": 3}[espec]
        pico = 2 if 8 <= hora <= 18 else -1
        ruido = np.random.randint(-1, 2)  # -1, 0, +1
        n = max(1, base + pico + ruido)
        return min(n, len(prof_por_espec[espec]))

    def lambda_atend(espec: str, hora: int) -> float:
        lam = BASE_LAMBDA[espec] * HOUR_MULT[hora]
        return max(0.15, lam)

    # -----------------------
    # Manchester: metas e ATP-alvo (para tempos de espera realistas)
    # -----------------------
    LIMITES  = {"Vermelho": 0, "Laranja": 10, "Amarelo": 60, "Verde": 120, "Azul": 240}  # min
    ATP_ALVO = {"Azul": 0.80, "Verde": 0.74, "Amarelo": 0.88, "Laranja": 0.82, "Vermelho": 0.00}

    def sorteia_manchester(espec: str) -> str:
        cores = ["Azul", "Verde", "Amarelo", "Laranja", "Vermelho"]
        if espec == "Cl√≠nica":   p = [0.15, 0.55, 0.22, 0.07, 0.01]
        elif espec == "Pediatria": p = [0.10, 0.50, 0.28, 0.10, 0.02]
        elif espec == "Cirurgia":  p = [0.05, 0.40, 0.35, 0.17, 0.03]
        else:                      p = [0.20, 0.60, 0.17, 0.03, 0.00]
        return random.choices(cores, weights=p, k=1)[0]

    # tempo de espera (chegada ‚Üí atendimento) bimodal por classe para atingir ATP aproximado
    def espera_minutos(classe: str) -> float:
        limite = LIMITES[classe]
        p_atp  = ATP_ALVO[classe]
        if classe == "Vermelho":
            return 1 + np.random.exponential(scale=2.0)  # sempre >0
        if np.random.rand() < p_atp:
            return float(np.random.triangular(left=0, mode=0.35*limite, right=0.9*limite))
        else:
            excess = np.random.exponential(scale=0.6*limite if limite > 0 else 8.0)
            return float(limite + 1 + excess)

    def atraso_triagem(classe: str) -> float:
        return np.random.uniform(0, 3) if classe == "Vermelho" else np.random.uniform(5, 20)

    # -----------------------
    # Gera√ß√£o principal
    # -----------------------
    inicio = pd.Timestamp("2025-05-01")
    registros = []
    atendimento_seq = 1

    for d in range(dias):
        data = inicio + pd.Timedelta(days=d)
        for hora in range(24):
            for espec in especialidades:
                n_prof = n_prof_ativos(espec, hora)
                profs_escala = random.sample(prof_por_espec[espec], n_prof)
                lam = lambda_atend(espec, hora)

                for prof in profs_escala:
                    n_atend = np.random.poisson(lam)
                    for _ in range(n_atend):
                        minuto, segundo = np.random.randint(0, 60), np.random.randint(0, 60)
                        dt_chegada = data + pd.Timedelta(hours=hora, minutes=minuto, seconds=segundo)

                        classe = sorteia_manchester(espec)
                        t_espera = espera_minutos(classe)
                        t_triag  = min(atraso_triagem(classe), max(t_espera - 1, 0))

                        dt_triagem = dt_chegada + pd.Timedelta(minutes=t_triag)
                        dt_atend   = dt_chegada + pd.Timedelta(minutes=t_espera)

                        registros.append({
                            "atendimento_id": f"ATD_{atendimento_seq:07d}",
                            "data_hora_chegada": dt_chegada,
                            "data_hora_triagem": dt_triagem,
                            "data_hora_atendimento": dt_atend,
                            "especialidade": espec,
                            "profissional_id": prof,
                            "hora": (dt_atend if usar_hora == "atendimento" else dt_chegada).hour,
                            "classificacao_risco": classe,
                        })
                        atendimento_seq += 1

    df = pd.DataFrame(registros).sort_values("data_hora_chegada").reset_index(drop=True)

    # -----------------------
    # Inje√ß√£o de OUTLIERS (picos)
    # -----------------------
    if outlier_days and outlier_hours:
        datas_unicas = pd.to_datetime(df["data_hora_chegada"].dt.date.unique())
        datas_escolhidas = np.random.choice(datas_unicas,
                                            size=min(outlier_days, len(datas_unicas)),
                                            replace=False)

        extras = []
        for dia in datas_escolhidas:
            for h in outlier_hours:
                for espec in especialidades:
                    # gera volume extra proporcional ao normal * outlier_factor
                    lam_extra = BASE_LAMBDA[espec] * HOUR_MULT[h] * (outlier_factor - 1.0)
                    n_extra = np.random.poisson(lam_extra * len(prof_por_espec[espec]))
                    for _ in range(n_extra):
                        minuto, segundo = np.random.randint(0, 60), np.random.randint(0, 60)
                        dt_chegada = pd.Timestamp(dia) + pd.Timedelta(hours=int(h),
                                                                      minutes=minuto, seconds=segundo)

                        classe = sorteia_manchester(espec)
                        t_espera = espera_minutos(classe)
                        t_triag  = min(atraso_triagem(classe), max(t_espera - 1, 0))

                        dt_triagem = dt_chegada + pd.Timedelta(minutes=t_triag)
                        dt_atend   = dt_chegada + pd.Timedelta(minutes=t_espera)

                        extras.append({
                            "atendimento_id": f"ATD_{atendimento_seq:07d}",
                            "data_hora_chegada": dt_chegada,
                            "data_hora_triagem": dt_triagem,
                            "data_hora_atendimento": dt_atend,
                            "especialidade": espec,
                            "profissional_id": random.choice(prof_por_espec[espec]),
                            "hora": (dt_atend if usar_hora == "atendimento" else dt_chegada).hour,
                            "classificacao_risco": classe,
                        })
                        atendimento_seq += 1

        if extras:
            df = pd.concat([df, pd.DataFrame(extras)], ignore_index=True)
            df = df.sort_values("data_hora_chegada").reset_index(drop=True)

    return df
```

```python
# -----------------------
# EXEMPLO DE USO
# -----------------------
df = gerar_dados_pa_com_outliers(
    dias=14,
    seed=42,
    outlier_hours=(10, 17),   # picos √†s 10h e 17h
    outlier_days=5,           # em 2 dias aleat√≥rios
    outlier_factor=2.8,       # for√ßa do pico
    usar_hora="atendimento"   # ou "chegada" se quiser refletir demanda
)
```

```python
def carta_controle_pronto_atendimento(df):
    """
    Carta de controle (PA: atendimentos/h) com:
    - Linha original + m√©dia m√≥vel
    - Faixas de ¬±1œÉ e ¬±2œÉ
    - Mediana geral e medianas por turno (madrugada/dia/noite)
    - Outliers > +2œÉ destacados e anotados
    - Grid major/minor, locators, formata√ß√£o HHh, legenda completa
    """

    # ========================
    # 0) Imports
    # ========================
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib.ticker import MultipleLocator, FuncFormatter
    from matplotlib.patches import Patch
    import matplotlib.patheffects as pe
    from matplotlib.transforms import offset_copy

    # Paleta de cores (amig√°vel para dalt√¥nicos)
    cor_base = "#0072B2"   # linha principal
    cor_mm   = "#000000"   # m√©dia m√≥vel
    cor_1s   = "#009E73"   # ¬±1œÉ (verde)
    cor_2s   = "#F0E442"   # ¬±2œÉ (amarelo)

    # ========================
    # 1) Figura e eixos
    # ========================
    fig, ax = plt.subplots(figsize=(10, 5))  # Cria figura de 10x5 polegadas

    # ========================
    # 2) Linha principal e m√©dia m√≥vel
    # ========================
    ax.plot(df["hora"], df["atendimentos"],
            marker="o", linewidth=2.2, color=cor_base,
            label="Atendimentos/h")  # S√©rie original

    # M√©dia m√≥vel centrada (window=3), min_periods=1 evita NaN nas bordas
    df["media_movel"] = (df["atendimentos"]
                         .rolling(window=3, center=True, min_periods=1)
                         .mean())

    ax.plot(df["hora"], df["media_movel"],
            linestyle="--", linewidth=1.2, color=cor_mm,
            label="M√©dia m√≥vel (3h)")  # S√©rie suavizada

    # ========================
    # 3) T√≠tulos e limites
    # ========================
    ax.set_title("Pronto Atendimento - Fluxo Hor√°rio",
                 fontsize=14, fontweight="bold", pad=30)
    ax.set_xlabel("Hora do dia", fontsize=12, labelpad=8)
    ax.set_ylabel("Quantidade de atendimentos", fontsize=12, labelpad=8)

    # Limites autom√°ticos com pequena folga vertical e horizontal
    pad_y = 0.07 * (df["atendimentos"].max() - df["atendimentos"].min())
    ax.set_ylim(max(0, df["atendimentos"].min() - pad_y),
                df["atendimentos"].max() + pad_y)
    ax.set_xlim(df["hora"].min() - 0.2, df["hora"].max() + 0.2)

    # ========================
    # 4) Locators e formata√ß√£o dos eixos
    # ========================
    ax.xaxis.set_major_locator(MultipleLocator(2))   # ticks maiores a cada 2h
    ax.xaxis.set_minor_locator(MultipleLocator(1))   # ticks menores a cada 1h
    ax.yaxis.set_major_locator(MultipleLocator(5))   # ticks maiores a cada 5 unidades
    ax.yaxis.set_minor_locator(MultipleLocator(1))   # ticks menores a cada 1 unidade

    # Formata o eixo X para mostrar "HHh" apenas nos inteiros
    ax.xaxis.set_major_formatter(
        FuncFormatter(lambda v, pos: f"{int(v):02d}h" if v == int(v) else "")
    )

    # ========================
    # 5) Grid equilibrado
    # ========================
    ax.grid(True, which="major", linestyle="--", alpha=0.35)  # mais vis√≠vel
    ax.grid(True, which="minor", linestyle=":",  alpha=0.15)  # mais suave

    # ========================
    # 6) Mediana geral
    # ========================
    mediana = float(np.median(df["atendimentos"]))
    ax.axhline(mediana, linestyle="-.", linewidth=1.2, color="tab:blue")  # linha da mediana
    ax.text(df["hora"].max() + 0.3, mediana, "Mediana Geral",
            ha="left", va="bottom", fontsize=9,
            path_effects=[pe.withStroke(linewidth=3, foreground="white", alpha=0.9)])  # contorno para legibilidade

    # ========================
    # 7) Medianas por turno
    # ========================
    turnos = [(0, 7, "Madrugada"), (7, 19, "Dia"), (19, 24, "Noite")]
    for x0, x1, nome in turnos:
        sub = df[(df["hora"] >= x0) & (df["hora"] < x1)]
        if sub.empty:
            continue
        med_t = float(sub["atendimentos"].median())
        ax.hlines(med_t, x0, x1, linestyles="--", linewidth=1.6,
                  colors="tab:gray", zorder=2)  # linha horizontal
        x_centro = (x0 + x1) / 2
        ax.text(x_centro, med_t, f"Mediana {nome}",
                ha="center", va="bottom", fontsize=9, color="tab:gray",
                path_effects=[pe.withStroke(linewidth=3, foreground="white", alpha=0.9)])

    # Marca in√≠cio e fim do turno diurno
    ax.axvline(7,  linestyle=":", linewidth=1.2, color="tab:gray")
    ax.axvline(19, linestyle=":", linewidth=1.2, color="tab:gray")
    ax.text(7,  ax.get_ylim()[1], "In√≠cio turno diurno", ha="center", va="bottom")
    ax.text(19, ax.get_ylim()[1], "Fim turno diurno",   ha="center", va="bottom")

    # ========================
    # 8) Faixas ¬±1œÉ e ¬±2œÉ
    # ========================
    media  = df["atendimentos"].mean()
    desvio = df["atendimentos"].std()
    lim_inf_1, lim_sup_1 = media - desvio,   media + desvio
    lim_inf_2, lim_sup_2 = media - 2*desvio, media + 2*desvio

    # ¬±2œÉ mais claro, ¬±1œÉ mais escuro
    ax.axhspan(lim_inf_2, lim_sup_2, alpha=0.06, color=cor_2s, zorder=0)
    ax.axhspan(lim_inf_1, lim_sup_1, alpha=0.15, color=cor_1s, zorder=1)

    # R√≥tulos laterais em eixo ‚Äúfantasma‚Äù (twinx) para n√£o poluir o principal
    axr = ax.twinx()
    axr.set_ylim(ax.get_ylim()); axr.set_yticks([]); axr.grid(False)
    for y, txt, va in [(lim_sup_2, "Limite sup. 2œÉ", "bottom"),
                       (lim_sup_1, "Limite sup. 1œÉ", "bottom"),
                       (lim_inf_1, "Limite inf. 1œÉ", "top"),
                       (lim_inf_2, "Limite inf. 2œÉ", "top")]:
        axr.text(df["hora"].max() + 0.3, y, txt, ha="left", va=va, fontsize=8)

    # ========================
    # 9) Outliers (> +2œÉ)
    # ========================
    outliers = df[df["atendimentos"] > lim_sup_2]
    ax.scatter(outliers["hora"], outliers["atendimentos"],
               s=80, facecolors="none", edgecolors="red",
               linewidths=2, zorder=6, label="> +2œÉ")

    # ========================
    # 10) R√≥tulos num√©ricos nos pontos
    # ========================
    for i, (xv, yv) in enumerate(zip(df["hora"], df["atendimentos"])):
        # Mostra r√≥tulo apenas a cada 2 pontos ou se for outlier
        if (i % 2 == 0) or (df.index[i] in outliers.index):
            ax.text(xv, yv + 0.6, f"{int(yv)}",
                    ha="center", va="bottom", fontsize=8,
                    zorder=5, clip_on=True)

    # ========================
    # 11) Legenda completa
    # ========================
    handles, labels = ax.get_legend_handles_labels()
    handles += [
        Patch(facecolor=cor_1s, alpha=0.15, label="¬±1œÉ"),
        Patch(facecolor=cor_2s, alpha=0.06, label="¬±2œÉ"),
    ]
    ax.legend(handles=handles, loc="upper left", bbox_to_anchor=(0.01, 0.99),
              fontsize=10, title="Legenda", frameon=True, ncol=1)

    # ========================
    # 12) Salvar e mostrar
    # ========================
    plt.savefig("pa_atendimentos_linha.png", dpi=300, bbox_inches="tight")
    plt.show()
```

```python
# ----- Executar -----
carta_controle_pronto_atendimento(media_atendimentos_por_hora)
```
