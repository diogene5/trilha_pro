# 📚 Fundamentos de visualização

## 📊 Anatomia de um Gráfico

Um gráfico possui várias partes:
- **Figure**: toda a imagem do gráfico
- **Axes**: os eixos do gráfico (pode haver mais de um)
- **Axis**: o eixo X e o eixo Y individuais
- **Ticks**: marcações dos eixos
- **Labels**: rótulos dos eixos
- **Title**: título do gráfico
- **Legend**: legenda de elementos traçados

## ✍️ Gramática da configuração de um Gráfico


Podemos pensar na configuração de um gráfico em programação como uma gramática visual:
- **Dados** → `x`, `y`
- **Estética** → `cor`, `formato`, `linha`, `marcador`
- **Geometria** → `plot()`, `scatter()`, `bar()`
- **Escala** → linear, log, etc.
- **Facetas** → `subplot`
- **Tema** → `style`, grid, fonte, etc.

## 🧠 Bibliotecas para DataViz em Python

A visualização de dados é uma etapa essencial no processo de análise, pois traduz números em narrativas visuais compreensíveis. O ecossistema Python oferece uma variedade robusta de bibliotecas para visualização, cada uma com propósitos, níveis de abstração e públicos diferentes.

| Biblioteca            | Nível       | Baseada em       | Interativa | Estática | Facilidade | Performance | Suporte a Web | Casos de Uso Ideais                                      |
| --------------------- | ----------- | ---------------- | ---------- | -------- | ---------- | ----------- | ------------- | -------------------------------------------------------- |
| **Matplotlib**        | Baixo nível | Nativo Python    | ❌          | ✅        | Média      | Alta        | ❌             | Gráficos básicos, controle fino, publicações científicas |
| **Seaborn**           | Médio nível | Matplotlib       | ❌          | ✅        | Alta       | Alta        | ❌             | Estatísticas, EDA, dados estruturados (DataFrames)       |
| **Plotly**            | Alto nível  | JavaScript       | ✅          | ✅        | Alta       | Média       | ✅             | Dashboards interativos, web apps, relatórios dinâmicos   |
| **Bokeh**             | Alto nível  | JavaScript       | ✅          | ✅        | Média      | Média       | ✅             | Web apps, dashboards, interações ricas                   |
| **Altair**            | Alto nível  | Vega-Lite        | ✅          | ✅        | Alta       | Média       | ✅             | Visualização declarativa, EDA rápido, ciência de dados   |
| **ggplot (plotnine)** | Médio nível | Matplotlib       | ❌          | ✅        | Alta       | Média       | ❌             | Visualização tipo ggplot2 (R), análises estatísticas     |
| **Holoviews**         | Alto nível  | Bokeh/Matplotlib | ✅          | ✅        | Alta       | Alta        | ✅             | Grandes volumes de dados, pipelines declarativos         |
| **Pygal**             | Alto nível  | SVG              | ✅          | ✅        | Alta       | Baixa       | ✅             | Gráficos SVG elegantes para web                          |
| **Dash**              | Aplicações  | Plotly           | ✅          | ✅        | Alta       | Média       | ✅             | Construção de dashboards com backend Python              |
| **Panel**             | Aplicações  | Bokeh/Holoviews  | ✅          | ✅        | Alta       | Alta        | ✅             | Apps com widgets, dashboards complexos                   |


Não existe uma única melhor biblioteca — existe a melhor para o seu contexto. Entender o ecossistema permite fazer escolhas inteligentes e comunicar dados com clareza, precisão e impacto visual.

| Contexto                        | Bibliotecas Recomendadas       |
| ------------------------------- | ------------------------------ |
| Publicações científicas         | `Matplotlib`, `Seaborn`        |
| Análise exploratória (EDA)      | `Seaborn`, `Altair`, `Plotly`  |
| Dashboards interativos          | `Plotly`, `Dash`, `Panel`      |
| Visualização em web estática    | `Pygal`, `Altair`, `Plotly`    |
| Grandes volumes de dados        | `Holoviews`, `Datashader`      |
| Familiaridade com ggplot2 do R  | `plotnine`                     |
| Aplicações analíticas completas | `Dash`, `Panel`, `Streamlit`\* |
