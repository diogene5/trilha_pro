🔹 Com value_counts() podemos ver que a coluna Sexo está com os valores meio estranhos, veja só

```python
df_receitas['cliente_sexo'].value_counts(dropna=False)
```

🔹 Você pode usar a função np.where() para substituir onde tem F por Feminino. É uma boa escolha de função quando queremos substituir valores com base em uma única condição

```python
df_receitas["cliente_sexo"].value_counts()
```

```python
df_receitas["cliente_sexo"] = np.where(df_receitas["cliente_sexo"] == 'F', "Feminino", df_receitas["cliente_sexo"])
```

🔹 No entanto, aqui precisamos substituir dois valores, porque também M precisa ser substituído por Masculino. Nesse caso, é possível utilizar np.select()

```python
# Definir condições
condicoes = [df_receitas["cliente_sexo"] == "F",
             df_receitas["cliente_sexo"] == "M"
             ]

# Definir os valores correspondentes para cada condição. Tem que ter um valor para cada condição
valores = ["Feminino", "Masculino"]

# Aplicar a substituição com np.select() com um valor default
df_receitas["cliente_sexo"] = np.select(condicoes, valores, default= df_receitas["cliente_sexo"])
```

🔹 Só que em programação, existem vários caminhos! Você ainda pode utilizar funções como replace()

```python
# Substituir diretamente valores específicos

df_receitas["cliente_sexo"] = df_receitas["cliente_sexo"].replace({"F": "Feminino", "M": "Masculino"})
```

🔹 Ou a função map(). Só tome o cuidado, pois se um valor não estiver no dicionário, ele será convertido para NaN.

```python
# Criar um dicionário de mapeamento e aplicar com map().

mapa_sexo = {"F": "Feminino", "M": "Masculino", "Feminino": "Feminino", "Masculino": "Masculino"}

df_receitas["cliente_sexo"] = df_receitas["cliente_sexo"].map(mapa_sexo)
```

🔹 Por fim, utilize a função apply(), para aplicar uma função customizada por você!

```python
# Função personalizada para padronizar a coluna

def padronizar_sexo(valor):
    if valor in ["F", "Feminino"]:
        return "Feminino"
    elif valor in ["M", "Masculino"]:
        return "Masculino"
    return valor #retorna o valor original caso não seja alterado

# Aplicando a função com apply()
df_receitas["cliente_sexo"] = df_receitas["cliente_sexo"].apply(padronizar_sexo)
```

```python
# Criando categorias de idade com pd.cut()

df_receitas["cliente_faixa_etaria"] = pd.cut(
    df_receitas["cliente_idade"],
    bins=[0, 17, 35, 60, 100],  # Limites das faixas etárias
    labels=["0-17", "18-35", "36-60", "61-100"]
)
```

🔹 Existem muitos caminhos. Poderíamos ter usado o np.select novamente para isso

```python
# Criando condições para as faixas etárias
condicoes = [
    df_receitas["cliente_idade"] < 18,
    (df_receitas["cliente_idade"] >= 18) & (df_receitas["cliente_idade"] <= 35),
    (df_receitas["cliente_idade"] > 35) & (df_receitas["cliente_idade"] <= 60),
    df_receitas["cliente_idade"] > 60
]

# 📌 Definindo os rótulos correspondentes
categorias = ["0-17", "18-35", "36-60", "61-100"]

# 📌 Criando a nova coluna com np.select()
df_receitas["cliente_faixa_etaria"] = np.select(condicoes, categorias, default="Desconhecido")
```

```python
df_receitas["cliente_faixa_etaria"].value_counts(dropna=False)
```

Isso é o que chamamos de **feature engeniier ou engenharia de features**, uma habilidade especialmente importante

### 💡 Aplicando funções

No entanto, pode ser que a criação de features necessite de algum cálculo mais complexo. Um método muito usado na manipulação de dados é aplicar uma função nos dados ou em parte deles, para isso o método <font color = "orange">**`.apply()`</font>** é muito utilizado.

Importante dizer que esse método **aplica uma determinada função em apenas uma coluna ou linha do DF.**

```python
# Função para categorizar a idade
def classificar_faixa_etaria(idade):
    if idade < 18:
        return "0-17"
    elif idade <= 35:
        return "18-35"
    elif idade <= 60:
        return "36-60"
    elif idade > 60:
        return "61-100"
    else:
        return "Desconhecido"
```

```python
# Aplicando a função com apply()
df_receitas["cliente_faixa_etaria"] = df_receitas["cliente_idade"].apply(classificar_faixa_etaria)
```

**Comparações**

| Critério                       | `np.select()`                                           | `apply()` com função                                    |
| ------------------------------ | ------------------------------------------------------- | ------------------------------------------------------- |
| 🔄 **Velocidade**              | ✅ Mais rápido para grandes volumes de dados             | ❌ Mais lento (loop interno por linha)                   |

| 🧩 **Flexibilidade**           | ❌ Limitado a expressões booleanas simples               | ✅ Pode usar qualquer lógica Python                      |
| 🛠️ **Engenharia de features** | ✅ Ideal para regras simples de classificação            | ✅ Ideal para regras complexas e condicionais encadeadas |
| 🔍 **Depuração e testes**      | ❌ Mais difícil de testar e isolar erros                 | ✅ Função independente facilita testes                   |

##### 🚩 **INNER JOIN**

A partir desses 2 DataFrames, podemos notar que existe uma coluna em comum - a coluna 'chave' - que relaciona os dois.
A nossa intenção é unir as informações de cada DF cuja 'chave' tenha o mesmo valor. Para isso, usaremos o <font color = 'orange'> inner join</font>.

>
Lembra de conjuntos? Esse comando é como a **intersecção** de dois conjuntos - apenas as informações que pertecem aos dois!

```python
pd.merge(esquerda, direita, on='chave', how='inner')

#Os únicos valores comuns nas chaves são 2, 3 e 4. Portanto, nosso merge só terá essas linhas.
```

##### 🚩 **LEFT OUTER JOIN**

A diferença dessa função para a anterior é que agora o programa vai pegar e retornar todas as chaves do DataFrame da esquerda, já no da direita, só vai retornar os valores que existirem no da esquera. Caso não exista, retornará com `NaN`.

```python
pd.merge(esquerda, direita, on='chave', how='left')
```

##### 🚩 **RIGHT OUTER JOIN**

Análogo ao `merge` anterior, a ideia desse é pegar as chaves do DF da direita, verificar quais chaves existem na esquerda e, assim, devolver os respectivos valores.

```python
pd.merge(esquerda, direita, on='chave', how='right')
```

***
<font color = 'grey'> Talvez você seja ligeiro(a) e já tenha entendido esse negócio de left e right, mas talvez não. E como nós somos muuuuuuuuito didáticos, vamos explicar:
  
<font color = 'grey'> **Como que o programa sabe quem é `left` e quem é `right`, ali como parâmetro do `how`?**
  
 <font color = 'grey'> É bem óbvio, mesmo: o DataFrame referenciado primeiro, dentro de `pd.merge`, está à esquerda do DF referênciado em seguida. Ou seja, o `how='left'` está se referindo ao primeiro DF (Por isso, no nosso exemplo, chamamos o DataFrame de 'esquerda')
   
<font color = 'grey'>Perceba que se trocarmos a ordem e colocarmos o DataFrame 'direita' à esquerda do DataFrame 'esquerda', teremos exatamente o contrário e, portanto, ao usar `how='left'` na verdade ele estará referenciando o DataFrame 'direita'
   
   ****
  

```python
pd.merge(direita, esquerda, on='chave', how='right')
```

##### **🚩 FULL OUTER JOIN**

Por fim, essa última forma de <font color = 'orange'> `join` </font>basicamente pega todas as chaves existentes em um **ou** outro DataFrame e, com isso, constrói um terceiro com todas elas, havendo conexão entre os DFs ou não.
>
Esse aqui pode ser comparado com a **união** de dois conjuntos!


```python
pd.merge(esquerda, direita, on='chave', how='outer')
```

##### **🚩 COMBINING**

Existe, ainda, uma forma de juntar esses 2 DataFrames sem fazer a conexão entre as chaves existentes em ambos. É como se você simplesmente "costurasse" um DataFrame no outro.

```python
pd.concat([esquerda, direita])
```

E você pode preferir "costurar" esses DFs de lado. Tudo bem! Olha só:

```python
pd.concat([esquerda, direita], axis=1)
```

### 💡 Agregações

🔹 Podemos utilizar a função groupby, muito conhecida do SQL. Passe as colunas pelas quais quer agregar (como dimensão). As colunas que quer agregar (como fato) e a forma de agregação

#### 🧐 Teoria

Presta atenção aqui! Para usar essa função, precisamos entender algumas coisinhas, pois o groupby() funciona em três etapas — pense como um processo de fábrica

| Etapa        | O que acontece                                                   | Exemplo                              |
| ------------ | ---------------------------------------------------------------- | ------------------------------------ |
| **Dividir**  | Divide os dados em grupos, com base em uma ou mais colunas       | Agrupar por sexo, curso, cidade...   |
| **Aplicar**  | Aplica uma função em cada grupo (ex: média, soma, contagem, etc) | Média de notas, soma de vendas, etc. |
| **Combinar** | Junta os resultados em uma nova tabela, organizada por grupo     | Mostra os resultados lado a lado     |


```python
import pandas as pd

# Criando a tabela (DataFrame)
dados = pd.DataFrame({
    'Aluno': ['Ana', 'Bruno', 'Carla', 'Daniel'],
    'Sexo': ['F', 'M', 'F', 'M'],
    'Turma':['Turma A', 'Turma A', 'Turma B', 'Turma B'],
    'Nota': [8.5, 7.0, 9.0, 6.5]
})
```

```python
dados.groupby('Sexo')['Nota'].mean()
```

Explicação:
* dados.groupby('Sexo')   # Agrupa as linhas por valores iguais na coluna 'Sexo'
* ['Nota']                # Escolhe a coluna numérica onde será feita a operação
* .mean()                 # Aplica a média a cada grupo

**Se você quiser agrupar por mais de uma coisa:**

```python
dados.groupby(['Sexo', 'Turma'])['Nota'].mean()
```

**Se você quiser calcular mais de uma medida:**

```python
dados.groupby("Sexo")["Nota"].agg(["mean", "min"])
```

```python
dados.groupby("Sexo")["Nota"].aggregate(["mean", "min"])
```

No Pandas, .agg() e .aggregate() são exatamente a mesma coisa. A diferença está somente no nome — ambas as funções fazem a mesma operação.

**Se você quiser agregar calculando uma função também é possível:**

👩‍🏫 Às vezes, queremos fazer mais do que só calcular uma média, soma ou contagem. A função .apply() permite aplicar qualquer lógica personalizada em cada grupo!

```python
def classificar_grupo(grupo):
    media = grupo["Nota"].mean()
    if media > 8:
        return "Excelente"
    elif media >= 6:
        return "Regular"
    else:
        return "Ruim"

# Aplicando por grupo de Sexo
dados.groupby("Sexo").apply(classificar_grupo)
```

### 💡 Stack & Unstack

Acabamos de aprender uma forma de organizar o DataFrame em relação ao seu índice, agora vamos dar uma olhadinha em outra maneira de modelar nosso `DataFrame`.

Os métodos <font color = "orange">**`.stack()`</font>** e <font color = "orange">**`.unstack()`</font>** podem ser usados tanto em `Series` quanto em `DataFrames` e são usados juntamente com o conceito de `Multi-index` que acabamos de ver.

#### `Stack`

Em português, "empilhar", é um método que, basicamente, empilha os dados do DataFrame em uma ou mais linhas. Ou seja, move colunas para o índice (linha) → deixa o DataFrame mais longo e estreito

```python
#Primeiro vamos criar um DataFrame

colunas = pd.MultiIndex.from_product([['2017', '2018'],['Vendas', 'Clientes']])
df_exemplo=([[12, 45, 67, 56],[78, 89, 45, 67],[45, 67, 89, 90],[67, 44, 56, 55]])


df_exemplo = pd.DataFrame(df_exemplo,
                  index=['Ana', 'Bruno', 'Cristina', 'Jackson'],
                  columns=colunas)
df_exemplo
```

Aqui temos uma estrutura de colunas com dois níveis:
* Nível 0: ano (2017, 2018)
* Nível 1: tipo de dado (Clientes, Vendas)

Vamos então transformar as colunas do nível 0 em linhas:

```python
empilhado = df_exemplo.stack(0)
empilhado
```

E agora vamos então transformar as colunas do nível 1 em linhas:

```python
empilhado = df_exemplo.stack(1)
empilhado
```

Se não passar o parâmetro, será adotado o nível interno (maior)

```python
empilhado = df_exemplo.stack()
empilhado
```

#### `Unstack`

É o método inverso ao `stack`, ou seja, ele pega os índices interiores de linhas e transforma em índices interiores de colunas. Ou seja, move linhas para colunas → deixa o DataFrame mais curto e longo.

```python
empilhado
```

Aqui temos uma estrutura de índices (linhas) com dois níveis:
* Nível 0: nomes
* Nível 1: tipo de dado (Clientes, Vendas)

Vamos então transformar as linhas do nível 0 em linhas:

```python
desempilhado = empilhado.unstack(0)
desempilhado
```

E agora vamos então transformar os índices do nível 1 em linhas:

```python
desempilhado = empilhado.unstack(1)
desempilhado
```

Se não passar o parâmetro, será adotado o nível interno (maior)

```python
desempilhado = empilhado.unstack()
desempilhado
```

Perceba que o "desempilhado" é o mesmo DataFrame que o original "df". Portanto, o `unstack` desfez o que o `stack` tinha feito, por isso esses dois métodos são como duas operações inversas.

Basicamente, usamos o `stack` quando queremos os dados agrupados em menos colunas e mais linhas, e, caso contrário, usamos o `unstack`.

Usando o conceito de `multi-index`, podemos támbem usar vários índices no `pivot table`:

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values='saldo'
)

sales_pivot
```

Aqui, podemos ver que como não indicamos o que queremos fazer com os dados, o método simplismente faz uma média dos valores das colunas, mas e se quisermos uma realizar uma operação diferente para as colunas?
Lembra da função `agg`? Usávamos com o groupby para realizar alguma operação nos dados agrupados. Aqui podemos usar o método `aggfunc`


```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values='saldo',
    aggfunc="sum",
)

sales_pivot
```

E se quisermos mais de uma função? Temos!

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values='saldo',
    aggfunc=['sum', 'mean']
)

sales_pivot
```

Ainda, se quisermos ver colunas diferentes e aplicar funções diferentes em cada uma? Temos!

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values=['movimentações', 'saldo'],
    aggfunc={'movimentações': 'sum','saldo': ['sum','max']}
)

sales_pivot
```

Além disso, lembra do unstack? Aqui também podemos "desempilhar" os dados, por exemplo, podemos selecionar dados que estavam em formato coluna e transformá-los em linhas, com o método columns:

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values=['movimentações', 'saldo'],
    columns=['tipo'],
    aggfunc={'movimentações': 'sum','saldo': ['sum','max']}
)

sales_pivot
```

Ficou meio ruim de visualizar com todos esses `NaNs` né? Vamos substitui-los com o método `fill_value`

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta'],
    values=['movimentações', 'saldo'],
    columns=['tipo'],
    aggfunc={'movimentações': 'sum','saldo': ['sum','max']},
    fill_value=0
)

sales_pivot
```

O que queremos mostrar é que o método `pivot table` é muito útil e versátil para visualizarmos os dados da maneira em que desejamos! Podemos brincar com ele e descobrir diversas maneiras eficientes para entendermos melhor os dados que temos.

Por exemplo, podemos mover outros itens para o `index` para termos diferentes modos de visualização. Como remover o 'tipo' de `columns` e movê-lo para o `index`

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta', 'tipo'],
    values=['movimentações', 'saldo'],
    aggfunc={'movimentações': 'sum','saldo': ['sum','max']},
    fill_value=0
)

sales_pivot
```

E se quisermos ver um total de tudo isso? Dá uma olhadinha nesse método: `margins=True`, ele nos mostra a soma das colunas de valores:


```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta', 'tipo'],
    values=['movimentações', 'saldo'],
    aggfunc='sum',
    fill_value=0,
    margins=True
)
sales_pivot
```

E mais, podemos dar um nome para essa nova linha com `margins_name`:

```python
sales_pivot = sales.pivot_table(
    index=['gerente', 'conta', 'tipo'],
    values=['movimentações', 'saldo'],
    aggfunc='sum',
    fill_value=0,
    margins=True,
    margins_name='Total'
)
sales_pivot
```

**Comparação Pivot Table e Groupby**

Conseguiu ver a diferença entre os métodos? O importante é saber que podemos obter informações de várias formas diferentes, por isso, para usar o Pandas é essencial saber claramente o que você está procurando, assim fica mais fácil escolher o método mais adequado.

Para resumir o que acabamos de discutir:

| Característica                        | `pivot_table` / `stack` / `unstack`                         | `groupby`                                      |
| ------------------------------------- | ----------------------------------------------------------- | ---------------------------------------------- |
| 🧠 **Objetivo**                       | Reorganizar, modelar e transformar a visualização dos dados | Resumir os dados por grupos                    |
| 📦 **Retém todos os dados?**          | Sim. Os dados não são perdidos, apenas reorganizados        | Não. Retorna apenas os dados agregados         |
| 📐 **Formato de saída**               | DataFrame reorganizado, geralmente com multi-índices        | DataFrame ou Series com menos informações      |
| 🔧 **Funcionalidade principal**       | Mudança de forma: colunas viram linhas, ou vice-versa       | Cálculo de agregações, estatísticas por grupo  |
| 🔁 **Reversibilidade**                | `stack/unstack` são reversíveis (`df.stack().unstack()`)    | Não reversível diretamente                     |
| 💡 **Uso típico**                     | Tabelas dinâmicas, dashboards, visualizações tabulares      | Sumários estatísticos, análise de agrupamentos |
| 🧮 **Agregações simultâneas**         | Sim, com `pivot_table(aggfunc={...})`                       | Sim, mas separadamente ou com `.agg({...})`    |
| 🧱 **Facilidade de múltiplos níveis** | Muito boa (MultiIndex nativo)                               | Também boa, mas mais manual                    |


🎯 Quando usar cada um?
* Use groupby quando quiser resumos estatísticos simples.
* Use pivot_table para análises multidimensionais, como em tabelas dinâmicas.
* Use stack e unstack para transformar a estrutura de linhas/colunas — ótimo para visualizações e pré-processamento.

```python
- index: Coluna que vai para o eixo das linhas (índice da tabela)
- columns: Coluna que será transformada em colunas
- values: Coluna cujos valores serão agregados
- aggfunc: Função de agregação (ex: 'mean', 'sum', 'count', etc.)
```
